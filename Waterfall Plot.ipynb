```python
from blimpy import Filterbank
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path_to_fil = ("spliced_blc0001020304050607_guppi_"
               "58100_80372_OUMUAMUA_OFF_0016.gpuspec.0002.fil")
```

```python
obs = Filterbank(path_to_fil)
```

```python
obs.info()
```

```python
plt.figure(figsize=(10,6))
obs.plot_waterfall()
title = plt.title(
    "Original Waterfall Plot (Unzoomed)")
```

```python
plt.figure(figsize=(10,6))
obs.plot_waterfall(f_start=1800, f_stop=3000)
title = plt.title(
    "Zoom in to frequencies at 1800-3000 MHz")
```

```python
plt.figure(figsize=(10,6))
obs.plot_spectrum()
title = plt.title(
    "Spectrum Graph at time 0")
```

```python
plt.figure(figsize=(10,6))
obs.plot_spectrum(f_start=1700, f_stop=2800)
title = plt.title(
    "Spectrum Graph at time 0, zoomed in to frequencies of 1700-2800 MHz")
```
