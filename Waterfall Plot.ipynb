```python
from blimpy import Filterbank
from blimpy_yuhong.blimpy import Filterbank as FilDev
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path_to_fil = ("./data/noise_with_dc_2.rawspec.0001.fil")
```

```python
obs = Filterbank(path_to_fil)
obsD = FilDev(path_to_fil)
```

```python
obs.info()
```

```python
plt.figure(figsize=(10,6))
obsD.plot_waterfall(cmap='magma')
title = plt.title(
    "Original Waterfall Plot (Unzoomed)")
```

```python
plt.figure(figsize=(10,6))
obs.plot_waterfall()
title = plt.title(
    "Original Waterfall Plot (Unzoomed)")
```

```python
plt.figure(figsize=(10,6))
obs.plot_waterfall(f_start=1800, f_stop=3000)
title = plt.title(
    "Zoom in to frequencies at 1800-3000 MHz")
```

```python
plt.figure(figsize=(10,6))
obs.plot_spectrum()
title = plt.title(
    "Spectrum Graph at time 0")
```

```python
plt.figure(figsize=(10,6))
obs.plot_spectrum(f_start=1700, f_stop=2800)
title = plt.title(
    "Spectrum Graph at time 0, zoomed in to frequencies of 1700-2800 MHz")
```
